```python
Hypothesis testing
It is assumed that the mean systolic blood pressure is μ = 120 mm Hg. In the Honolulu Heart Study, a sample of n = 100 people had an average systolic blood pressure of 130.1 mm Hg with a standard deviation of 21.21 mm Hg. Is the group significantly different (with respect to systolic blood pressure!) from the regular population?

1. Set up the hypothesis test.
2. Write down all the steps followed for setting up the test.
3. Calculate the test statistic by hand and also code it in Python. It should be 4.76190. We will take a look at how to make decisions based on this calculated value.
𝑡=(𝑋¯−𝜇0)𝜎̂ /𝑛√ 
where:

𝑋¯  is the sample mean
𝜇0  is the population mean
𝜎̂   is the sample standard deviation
𝑛  is the number of measures in our sample
```

```python
𝑋¯ = 130.1 mm
𝜇0 = 120 mm Hg
𝜎̂ = 21.21 mm Hg
n = 100
```

1. Set up the hypothesis test. 

H0 = Sample mean is equal to the population mean.  
Question = Is the group significantly different from the regular population ?

One statistic is equal to another, that's our null hypothesis. So, we are going to do a 2 tail test, 97.5% confidence in a standard normal distribution

2. Write down all the steps followed for setting up the test.


```python
#1. Fisrt we import the library
import math

# 2. Second, we determine the variables and run python in order to obtain t

sample_mean = 130.1
pop_mean = 120
sample_std = 21.21
n = 100
statistic = (sample_mean - pop_mean)/(sample_std/math.sqrt(n))
print("Statistic is: ", statistic)

# 3. The t-statistic is equal to 4.76
```

Now that we have the t-test value, we need to choose which table are we following: 

- As we don't know the standard deviation of the population we will go for the t statistic
- Degree of freedoms = 100 - 1 = 99
- The t-tble (https://www.sjsu.edu/faculty/gerstman/StatPrimer/t-table.pdf), for 99 df and 0.025 per tail = 1.984

The critical region approach tells us to reject the null hypothesis at the 0.05 level if t>or= 0.025,99 = 1,984 or if t<=0.025,99 = -1,984 (We look at the table of Standard Normal Distribution to obtain 1,96). Therefore, we reject the null hypothesis because 4,76 is bigger than 1,984, therefore falls in the rejection region:

Rejecting H0, we accept H1 as true, so the sample mean is significantly different to the regular population

Before we apply PCA we need to standardize/scale the data. For this we can use sklearn's preprocessing library 

```python
from sklearn.preprocessing import StandardScaler
transformer = StandardScaler().fit(numerics)
scaled = transformer.transform(numerics)
scaled = pd.DataFrame(scaled)
scaled.columns = numerics.columns 
scaled.head()
```

```python
from sklearn.decomposition import PCA
from sklearn import preprocessing
pca = PCA()
pca.fit(scaled)
```

```python
# variations = pca.explained_variance_
variations = pca.explained_variance_ratio_*100
variations = np.round(variations, decimals=1)
variations
```
